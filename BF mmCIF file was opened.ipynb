```python
parser = MMCIFParser()

#The first argument to the instance method get_structure is an optional name for the molecule, and the second argument is the path to the structure file.
```

```python
# Extraction of all the structural information from the file.

structure = parser.get_structure('6DBP', 'db/6DBP.cif')
```

```python
#In order to see attributes of the obj we can run this code 
def cleandir(obj):
    print(", ".join([a for a in dir(obj) if not a.startswith("_")]))
cleandir(structure)
```

```python
for model in structure:
    print(f"model {model}")
```

```python
model = structure[0]
for chain in model:
    print(f"chain{chain}, Chain ID: {chain.id}")
```

```python
chain_A = model['A']
```

```python
for res in chain_A:
    print(f"Residue name: {res.resname}, number: {res.id[1]}")
```

```python
#Now we can retrieve any residue
res = chain_A[40]
print(res)
```

```python
#We go through all atoms in give residue
for atom in res:
    print(f"{atom.name}")
```

```python
for model in structure:
    for chain in model:
        for residue in chain:
            for atom in residue:
                print(atom)
```
